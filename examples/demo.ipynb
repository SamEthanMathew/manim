# JSON to Teaching Syllabus Converter Demo

This notebook demonstrates how to use the JSON to Teaching Syllabus converter for your Manim video pipeline.


```python
from json_to_syllabus import SyllabusGenerator, convert_json_to_syllabus
import json
from pprint import pprint
```

## Example 1: Simple Course Syllabus


```python
# Create example JSON data
example_data = {
    "document_type": "syllabus",
    "index": ["Introduction", "Key Concepts", "Summary"],
    "sections": [
        {
            "id": 0,
            "title": "Introduction",
            "content": {
                "topic": "Machine Learning Basics",
                "duration": "5 minutes",
                "objectives": ["Understand ML fundamentals", "Learn key terminology"]
            }
        },
        {
            "id": 1,
            "title": "Key Concepts",
            "content": {
                "supervised_learning": "Learning from labeled data",
                "unsupervised_learning": "Finding patterns in unlabeled data",
                "reinforcement_learning": "Learning through rewards"
            }
        },
        {
            "id": 2,
            "title": "Summary",
            "content": "Review the three main types of machine learning"
        }
    ]
}

print("Input JSON:")
pprint(example_data)
```

```python
# Convert to teaching syllabus
syllabus = convert_json_to_syllabus(json.dumps(example_data))
print(syllabus)
```

## Example 2: Load from JSON file


```python
# Load from the example file and save output
syllabus_from_file = convert_json_to_syllabus('example_input.json', 'demo_output.md')
print("✅ Syllabus generated and saved to demo_output.md")
print("\nFirst 500 characters:")
print(syllabus_from_file[:500])
```

## Example 3: Using the SyllabusGenerator class directly


```python
# Use the class for more control
generator = SyllabusGenerator(example_data)

# Access individual components
print("Document Type:", generator.document_type)
print("Number of sections:", len(generator.sections))
print("\nTable of Contents:")
print(generator._generate_toc())
```
